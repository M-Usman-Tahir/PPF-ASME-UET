### Acessing the data

```python
# accessing columns
data[['Name', 'Type 1']]
```

```python
# Difference between .loc()  &  .iloc()
```

#### loc
- Takes index of rows
- Doesn't stop before the stopping value but at it  
- Takes boolean DataFrame and show all the row/col where True

```python
display(data.loc[0:1])
```

```python
display(data['Type 1']=='Grass')
print(data.loc[data['Type 1']=='Grass'].shape)
data.loc[data['Type 1']=='Grass']
```

#### iloc
- Takes index for location (rows, Col)
- Stop right before the stopping value
- Donot take booleans

```python
data.iloc[0:1]
```

```python
data.iloc[0:5, 3:8]
```

```python
data.iloc[0,5]
```

```python
# changing values of elements
data.iloc[0,5] = 50
```

```python
data.iloc[0:1]
```

```python
# adding/creating columns
# data['Total'] = data['HP'] + data['Attack'] ...
data['Total'] = data.iloc[:, 4:-1].sum(axis=1)
```

```python
data
```

```python
data.Total.isnull().sum()
```

### Making our own DataFrame

```python
MyDF = pd.DataFrame()
```

```python
MyDF  # empty DataFrame
```

```python
columns = ['Name', 'Age', 'pswrd']
MyDF2 = pd.DataFrame(columns=columns)
MyDF2
```

```python
rows = [['Ali', 20, '12qw'], ['Saqib', 21, 'asdfg']]
MyDF3 = pd.DataFrame(rows)
MyDF3
```

```python
MyDF4 = pd.DataFrame(rows, columns= columns)     # Most important
MyDF4
```

```python
index = ['index 1', 'index 2']
MyDF5 = pd.DataFrame(rows, columns= columns, index=index)
MyDF5
```

```python
dic = {'Name':['Ali', 'Saqib'], 'Age':[20, 21], 'pswrd':[123, 345]}    # Most Important
MyDF6 = pd.DataFrame(dic)
MyDF6
```

```python
newDF = pd.DataFrame({'Name':['Saad', 'Aslam'], 'Age':[20, 19], 'pswrd':[90, 98], 'New':['a', 'b']})
newDF
```

```python
MyDF6 = pd.concat([MyDF6, newDF], ignore_index=True)
```

```python
MyDF6
```

```python
new2 = pd.DataFrame({'New2':[1,2,3,4]})
new2
```

```python
a = pd.concat([MyDF6, new2], axis=1)
```
